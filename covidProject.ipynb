```python
import pandas as pd

# Loading the dataset
df = pd.read_csv('owid-covid-data.csv')

# View column names
print("Columns in the dataset:")
print(df.columns)

# Preview the first few rows
df.head()
```

```python
# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# Filter for selected countries
countries = ['Kenya', 'United States', 'India']
df = df[df['location'].isin(countries)]
```

```python
# Drop rows with missing date or critical values
df = df.dropna(subset=['date', 'total_cases', 'total_deaths'])
```

```python
# Convert date to datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
# Interpolate missing numeric values
numeric_cols = ['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'total_vaccinations']
df[numeric_cols] = df[numeric_cols].interpolate()

df.head()
```
